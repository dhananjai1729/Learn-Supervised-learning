```python
def get_best(columnname):
  temp5 = pd.DataFrame()
  flag =0
  # treating categorical and non-categorical variables separately
  try:
      temp5[columnname] = pd.to_numeric(X_temp2[columnname],errors = 'raise')
      flag=0
  except:
      temp5[columnname] = X_temp2[columnname]
      flag=1
  if flag==0:
    # Sort the column in descending order to get the highest values at the top
    sorted_column = temp5[columnname].sort_values(ascending=False)

    # Calculate the total number of rows in the column
    total_rows = len(sorted_column)

    # Calculate the number of rows that represent 10% of the total rows
    num_top_rows = int(np.ceil(total_rows * 0.1))

    # Select the top rows
    top_rows = sorted_column[:num_top_rows]

    # Take note of the lowest and highest values in the selection
    lowest_value = top_rows.min()
    highest_value = top_rows.max()
    return [lowest_value,highest_value]
  elif flag==1:
    answerscount = temp5[columnname].value_counts()
    return answerscount.keys()[0]
```

```python
def get_prompt(columnname):
  # personality
  if columnname.find('personality')!=-1:
    count = 0
    for x,y in zip(personality['assessmentType'],personality['order']):
      count+=1
      if (str(x)+'_'+str(y)) == columnname:
        return personality['prompt'].iloc[count]
  # voice
  elif columnname.find('voice')!=-1:
    for x,y in enumerate(voice['name']):
      if(columnname.find(y)!=-1):
        return voice['prompt'].iloc[x]
  # Grammar
  elif columnname.find('englishGrammar')!=-1:
    for x,y in enumerate(grammar['name']):
      if y == columnname:
        return grammar['prompt'][x]
  # Culture
  elif columnname.find('culture')!=-1:
    if re.search(r"culture_[012345].score.[abcd]",columnname):
      return culture[culturedict2[columnname[16]]][culture['Factor']==culturedict1[str(columnname[8])]]
  else:
    return 'No Prompt'
```

```python
def get_score(value):
  if value == 'Strongly Disagree':
    temp9= 1
  elif value == 'Disagree':
    temp9 = 10
  elif value == 'Neutral':
    temp9 = 100
  elif value == 'Agree':
    temp9 = 100
  elif value == 'Strongly Agree':
    temp9 =1000
  elif value == 'None':
    temp9= 1000000
  return temp9
```

```python
def get_feature_fit(columnname,value):
  temp5 = pd.DataFrame()
  flag =0
  # treating categorical and non-categorical variables separately
  try:
      temp5[columnname] = pd.to_numeric(X_temp2[columnname],errors = 'raise')
      flag=0
  except:
      temp5[columnname] = X_temp2[columnname]
      flag=1
  if flag==0:
    # Sort the column in descending order to get the highest values at the top
    sorted_column = temp5[columnname].sort_values(ascending=False)

    # Calculate the total number of rows in the column
    total_rows = len(sorted_column)

    # Calculate the number of rows that represent 10% of the total rows
    num_top_rows = int(np.ceil(total_rows * 0.1))

    # Select the top rows
    top_rows = sorted_column[:num_top_rows]

    # Take note of the lowest and highest values in the selection
    lowest_value = top_rows.min()
    highest_value = top_rows.max()
    fit = value/highest_value
    if fit>=0.75:
      fitness = 'high'
    elif fit>=0.5:
      fitness = 'medium'
    else:
      fitness = 'low'
    return fitness
  elif flag==1:
    answerscount = temp5[columnname].value_counts()
    if columnname.find('personality')!=-1:
      val1 = get_score(value)
      val2 = get_score(get_best(columnname))
      val3 = max(val1,val2)/min(val1,val2)
      if val3>=1000:
        return 'low'
      elif val3>=100:
        return 'medium'
      elif val3>=10:
        return 'good'
      else:
        return 'best'
    elif value == answerscount.keys()[0]:
      return 'high'
    else:
      return 'low'
```
